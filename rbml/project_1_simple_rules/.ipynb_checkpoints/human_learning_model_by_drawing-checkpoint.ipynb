# Human Learn Rule Based

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Load Data into 2 DF
# DF_1: X (all attributes)
# DF_2: y (target or Label to predict)
X,y= load_iris(return_X_y=True, as_frame=True)

# X has 4x attributes
# manually set the X.column attributes
X.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
# Lets inspect X and y
# Get the number of attributes in X
X.columns   # ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)'
# X.size      # get the X size    (150X4)--> (r X c)
# X.head(10)  # Display the first 10 rows
#
# # Get the number of attributes in y
# y.size      # 150 --> 150 rows
# y.head(10)  # Display the first 10 rows

```

```python
# Split the Dataset into Train Data and Test Data
'''
X_train -> pd.DF        --> 4x attributes
X_test  -> pd.DF        --> 4x attributes
y_train -> pd.DF        --> target Label
y_test  -> pd.DF        --> target Label
'''
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
# Prepare the final Training Dataset by concatenating X_train and y_train
train = pd.concat([X_train, pd.DataFrame(y_train)], axis=1)
```
